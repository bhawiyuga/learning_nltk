# Word Frequency 

## Frequency Distribution 

Counting frequency distribution of word

```python
from nltk import FreqDist
```

```python
text = "Saya ibu budi. Ibu budi suka makan nasi." 
```

```python
token = nltk.word_tokenize(text)
```

```python
lower_token = [t.lower() for t in token]
print lower_token
```

```python
fdist=FreqDist(lower_token)
```

```python
fdist.keys()
```

```python
for k in fdist.keys():
    print k+" "+str(fdist[k])
```
